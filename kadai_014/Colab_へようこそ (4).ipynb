```
# リストの作成
array = [
    "月曜日は晴れです",
    "火曜日は雨です",
    "水曜日は晴れです",
    "木曜日は晴れです",
    "金曜日は曇りです",
    "土曜日は曇りのち雨です",
    "日曜日は雷雨です"
]

# ディクショナリの作成
dictionary = {
    "mon": "晴れ",
    "tue": "雨",
    "wed": "晴れ",
    "thu": "晴れ",
    "fri": "曇り",
    "sat": "曇りのち雨",
    "sun": "雷雨"
}

# リストから水曜日を取り出して出力
print(array[2])

# ディクショナリから水曜日を取り出して出力
print("水曜日は" + dictionary["wed"] + "です")
```

```
var = 15

if var % 3 == 0 and var % 5 == 0:
    print("FizzBuzz")
elif var % 3 == 0:
    print("Fizz")
elif var % 5 == 0:
    print("Buzz")
else:
    print(var)
```

```
array = ["水","金","地","火","木","土","天","海","冥"]

# forループ
for planet in array:
    print(planet)

# whileループ
i = 0
while i < len(array):
    print(array[i])
    i += 1
```

```
# 消費税を計算する関数
def calc_tax(price, tax_rate):
    result = price + price * tax_rate / 100
    return result

# 関数を呼び出す（商品金額110円、消費税10%）
total_price = calc_tax(110, 10)

# 結果を表示
print(total_price)
```
